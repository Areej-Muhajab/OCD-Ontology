# **Survey data **

- The mean of all liker scale value (agree, disagree, not sure).
Mean or median score analysis: This involves assigning a numerical value to each response category (e.g., 1 for "disagree," 2 for "agree," 3 for "not sure") and then calculating the mean or median score for each question. This can give you a sense of the overall level of agreement or disagreement with each statement.
Factor analysis: This involves looking for underlying factors or themes that may be driving the responses to certain questions. For example, if several questions (3 first question) are related to (Intrusive thought definition), you could perform a factor analysis to see if there are any underlying factors (e.g.,emotion, functional impairment, time) that are associated with the responses).

# **The sum, mean, median of (Agree, disagree, not sure) for positive and negative statements.**

```python
posStatment.loc['sum'] = posStatment.sum()
negStatment.loc['sum'] = negStatment.sum()
```
